## Transform Data for Drivers and added 'age'

```python
 from pyspark.sql.functions import floor, datediff, current_date, col
 df_drivers= df_drivers.withColumn('age', floor(datediff(current_date(), col('dob')) / 365 ))


display(df_drivers)
```

```python
df_lap_drivers = df_laptimes.join(df_drivers, df_laptimes.driverId == df_drivers.driverId)
display(df_lap_drivers)
```

```python
from pyspark.sql.functions import avg
df_lap_drivers = df_lap_drivers.groupby('age').agg(avg('lap').alias('avg_lap'))

display(df_lap_drivers)
```

```python
df_lap_drivers.write.mode("overwrite").csv("/Volumes/gr5069/ne2388/inclass/lap_drivers.csv", header = True)
```
